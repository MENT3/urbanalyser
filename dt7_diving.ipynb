```python
meta = pd.read_csv("./row_data/meta_dossier_complet.csv", sep=";")
meta
```

# Match col names

```python
col_names = meta[["COD_VAR", "LIB_VAR"]].drop_duplicates(subset="COD_VAR").to_dict("index")
col_names = {aa["COD_VAR"]: aa["LIB_VAR"] for aa in col_names.values()}
```

```python
with open("./output/col_list", "w") as f:
  f.write("\n".join(col_names.values()))
```

```python
df = df.rename(columns=col_names)
```
